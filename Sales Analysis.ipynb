# Exploratory Data Analysis (EDA)

### Sales by Country

```python
sales_by_city = sales_data.groupby('COUNTRY')['SALES'].sum().sort_values(ascending = False)
plt.figure(figsize = (10,5))
sales_by_city.plot(kind = 'bar', color = 'skyblue')
plt.title('Total Sales by Country')
plt.ylabel('Total Sales')
plt.xticks(rotation = 45)
plt.show()
```

### Sales by Product Line

```python
plt.figure(figsize = (10,5))
product_sales = sales_data.groupby('PRODUCTLINE')['SALES'].sum().sort_values(ascending=False)
sns.barplot(x = product_sales.index, y = product_sales.values)
plt.title('Total Sales by Product Line')
plt.ylabel('Total Sales')
plt.xticks(rotation = 45)
plt.show()
```

### Monthly Sales Trend

```python
sales_data['MONTH'] = sales_data['ORDERDATE'].dt.month_name()
sales_data
```

```python
monthly_sales = sales_data.groupby('MONTH')['SALES'].sum()
monthly_sales
```

```python
monthly_sales.plot(kind = 'bar', color = 'lightgreen', figsize = (10,5))
plt.title('Total Sales by Month')
plt.ylabel('Total Sales')
plt.xlabel('Month')
plt.xticks(rotation = 45)
plt.show()
```

### Correlation Matrix

```python
sns.heatmap(sales_data[['QUANTITYORDERED','MSRP','PRICEEACH','SALES']].corr(), annot = True, cmap = 'Blues')
plt.title('Corelation between numeric variable')
plt.show()
```

### Insights Summary

```python
print("Key Insights:")
print("- Highest sales city:", sales_by_city.idxmax())
print("- Top-selling product line:", product_sales.idxmax())
print("- Average order value:", round(avg_sales,2))
```
